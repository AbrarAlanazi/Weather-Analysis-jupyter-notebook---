```python
import pandas as pd
import seaborn as sns
sns.set(color_codes=True)

import warnings
warnings.filterwarnings('ignore')
```

```python
weather = pd.read_csv('weather.csv')
```

```python
weather.head()
```

```python
weather.info()
```

```python
sns.barplot(data = weather , x='humidity',  y='temperature')
```

```python
sns.displot(weather['humidity'])
```

```python
sns.displot(weather['humidity'], kde=False, rug=True);
```

```python
sns.jointplot(data = weather , x='humidity',  y='temperature')
```

```python
sns.jointplot(data = weather , x='humidity',  y='temperature' , kind="hex")
```

```python
sns.jointplot(data = weather , x='humidity',  y='temperature' , kind="kde")
```

```python
sns.pairplot(weather[['humidity', 'temperature', 'air_pollution_index']])
```

```python
sns.stripplot( data = weather , x = 'weather_type', y = 'temperature')
```

```python
sns.stripplot( data = weather , x = 'weather_type', y = 'temperature', jitter = True)
```

```python
sns.swarmplot( data = weather , x = 'weather_type', y = 'temperature')
```

```python
sns.boxplot( data = weather , x = 'weather_type', y = 'temperature', hue=weather['weather_type'])
```

```python
sns.barplot( data = weather , x = 'weather_type', y = 'temperature' , hue=weather['weather_type'])
```

```python
sns.countplot(data = weather , x = 'weather_type')
```

```python
sns.pointplot( data = weather , x = 'weather_type', y = 'temperature', hue=weather['weather_type'])
```

```python
sns.lmplot(x="humidity", y="temperature", data=weather)
```
